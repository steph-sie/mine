# Tracking ID Investigation

This notebook explores how YOLOv8 `track` output structures the tracking IDs and why they may not appear in the dashboard.

```python
from ultralytics import YOLO
import numpy as np
import cv2

model = YOLO('yolov8s.pt')
# attempt to get a real frame from sample video
cap = cv2.VideoCapture('test2.mp4')
ret, frame = cap.read()
cap.release()
if frame is None:
    frame = np.zeros((320, 320, 3), dtype=np.uint8)

# run tracking once
results = model.track(frame)
print('track result length', len(results))
if results:
    r = results[0]
    print('boxes data', r.boxes.data)
    print('boxes id attr', r.boxes.id)
    print('is_track flag', r.boxes.is_track)
    for b in r.boxes:
        print('box attributes', dir(b))
        try:
            print('vars', vars(b))
        except:
            pass

# run again to see ID persistence
results2 = model.track(frame)
print('second track length', len(results2))
if results2:
    r2 = results2[0]
    print('second boxes id', r2.boxes.id)
```
